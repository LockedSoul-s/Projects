import numpy as np
import pandas as pd
import gc

def generate_rounded_interval_features(df, initial_features):
    # 1. Round and convert to int for memory-efficient binning
    print("Rounding and binning data...")
    matrix = np.round(df[initial_features].values).astype(np.int16)
    
    unique_bins = np.unique(matrix)
    print(f"Found {len(unique_bins)} unique rounded intervals.")
    
    if len(unique_bins) > 400:
        print("⚠️ Warning: Too many intervals. Limiting to top 400.")
        bins, counts = np.unique(matrix, return_counts=True)
        unique_bins = bins[np.argsort(-counts)[:400]]

    new_cols = {}
    
    # 2. Re-adding the missing Bulk Stats
    raw_values = df[initial_features].values
    new_cols['fsum'] = raw_values.sum(axis=1)
    new_cols['fstd'] = raw_values.std(axis=1)
    new_cols['fskew'] = df[initial_features].skew(axis=1).values
    new_cols['fkurtosis'] = df[initial_features].kurtosis(axis=1).values # Added back
    
    # special1 logic: checks if the sum is between 72 and 75 (inclusive)
    #new_cols['special1'] = np.isin(new_cols['fsum'], [72, 73, 74, 75]) # Added back

    # 3. Quantiles
    for q in [0.0, 0.2, 0.4, 0.5, 0.6, 0.8, 1.0]:
        new_cols[f'q_{int(q*100)}'] = np.quantile(raw_values, q, axis=1)

    # 4. Interval Counts
    print("Calculating interval counts...")
    for v in unique_bins:
        new_cols[f'cnt_round_{v}'] = (matrix == v).sum(axis=1).astype(np.uint8)

    # 5. Final Concat
    new_df = pd.DataFrame(new_cols, index=df.index)
    df = pd.concat([df, new_df], axis=1)
    
    del matrix, raw_values, new_cols, new_df
    gc.collect()
    
    return df

```python
def enhance_features(df,features):
   
    df['sum_risk'] = df[features].sum(axis=1)
    df['mean_risk'] = df[features].mean(axis=1)
    df['std_risk'] = df[features].std(axis=1)
    df['max_risk'] = df[features].max(axis=1)
    df['min_risk'] = df[features].min(axis=1)
    df['median_risk'] = df[features].median(axis=1)
    df['f_median'] = df[features].median(axis=1)
    quantile_25 = df[features].quantile(0.25, axis=1)
    quantile_75 = df[features].quantile(0.75, axis=1)
    df['f_iqr_sum'] = df[features].apply(
        lambda x: x[(x >= x.quantile(0.25)) & (x <= x.quantile(0.75))].sum(), 
        axis=1
    )
    sorted_data = np.sort(df[features].values, axis=1)
    df['golden_sum'] = sorted_data[:, 0:15].sum(axis=1)
    df['peak_mean'] = sorted_data[:, 5:10].mean(axis=1)
    df['tail_sum'] = sorted_data[:, 15:].sum(axis=1)
    df['risk_consistency'] = df['golden_sum'] / (df['tail_sum'] + 1)
    df['f_iqr_range'] = quantile_75 - quantile_25
    df['f_trimean'] = (quantile_25 + (2 * df['f_median']) + quantile_75) / 4
    new_cols = {}
    # Interaction: Special characters of the synthetic generator 
    # often respond to unique value counts per row
    #df['unique_values'] = df.apply(lambda x: len(np.unique(x)), axis=1)
    print("Rounding and binning data...")
    matrix = np.round(df[initial_features].values).astype(np.int16)
    
    unique_bins = np.unique(matrix)
    print(f"Found {len(unique_bins)} unique rounded intervals.")
    
    if len(unique_bins) > 400:
        print(" Warning: Too many intervals. Limiting to top 400.")
        bins, counts = np.unique(matrix, return_counts=True)
        unique_bins = bins[np.argsort(-counts)[:400]]
    print("Calculating interval counts...")
    for v in unique_bins:
        new_cols[f'cnt_round_{v}'] = (matrix == v).sum(axis=1).astype(np.uint8)  
    new_df = pd.DataFrame(new_cols, index=df.index)
    df = pd.concat([df, new_df], axis=1)
    
    
    return df
print("--- Processing Train ---")
target=train_data['FloodProbability']
train_data_yap=train_data.drop(columns=['FloodProbability'],axis=1)
train_data_yap = enhance_features(train_data_yap, initial_features)
train_data_yap['FloodProbability']=target

print("\n--- Processing Test ---")
test_data_yap = enhance_features(test_data.copy(), initial_features)
```
